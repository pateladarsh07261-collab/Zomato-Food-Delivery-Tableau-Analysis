```python
import pandas as pd
input_file = "zomato.csv"  
columns_needed = ['name', 'online_order', 'book_table', 'rate', 'votes', 'location', 'rest_type', 'approx_cost(for two people)', 'listed_in(type)']
df = pd.read_csv(input_file, usecols=columns_needed, nrows=10000)

```

```python
df.head(5)
```

```python
df.columns = ['restaurant_name', 'online_order', 'book_table', 'rate_out_of_5', 'total_votes', 'location', 'rest_type', 'approx_cost_for_two', 'type_of_meal']
```

```python
df['rate_out_of_5'] = df['rate_out_of_5'].str.split('/').str[0].str.strip()
```

```python
df['rate_out_of_5'] = pd.to_numeric(df['rate_out_of_5'], errors='coerce')
```

```python
df.to_csv("zomato_clean_small.csv", index=False)
```
